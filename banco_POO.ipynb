```python
from datetime import datetime

class Transacao:
    @property
    def __init__(self) -> None:
        pass

    def registrar(self,conta):
        return

class Deposito(Transacao):
    def __init__(self,valor) -> None:
        super().__init__()
        self.valor = valor

    def registrar(self,conta):
        if conta.depositar(self.valor):
            conta.historico.adicionar_transacao(self)
            return(True)
        return(False)

class Saque(Transacao):
    def __init__(self,valor) -> None:
        super().__init__()
        self.valor = valor

    def registrar(self,conta):
        if conta.sacar(self.valor):
            conta.historico.adicionar_transacao(self)
            return(True)
        return(False)

class Historico:
    def __init__(self):
       self.lista_historico = []

    @property
    def lista_historico(self):
        return self.lista_historico

    def adicionar_transacao(self,transacao):
        string_date = datetime.now()
        string_date = string_date.strftime("%d/%m/%Y %H:%M:%S")
        registro = {"transaction_type": transacao.__class__.__name__, "date":string_date, "value":transacao.valor}
        self.lista_historico.apprend(registro)

class Cliente:
    def __init__(self,endereco):
        self.endereco = endereco
        self.contas = []

    def add_conta(self,conta):
        self.contas.append(conta)

    def realizar_transacao(conta, transacao):
        transacao.lower()
        if transacao.lower() == 'deposito':
            conta.depositar()
        if transacao.lower() == 'saque':
            conta.sacar()
        if transacao.lower() == 'saldo':
            conta.saldo()
        return

class PessoaFisica(Cliente):
    def __init__(self, endereco,cpf,nome,data_nascimento):
        super().__init__(endereco)
        self.cpf = cpf
        self.nome = nome
        self.data_nascimento = data_nascimento


class Conta:
    def __init__(self,numero,cliente):
        self.saldo = 0
        self.numero = numero
        self.agencia = "0001"
        self.cliente = cliente
        self.historico = Historico()

    @property
    def numero(self):
      return self.saldo

    @property
    def numero(self):
      return self.numero

    @property
    def numero(self):
      return self.agencia

    @property
    def numero(self):
      return self.cliente


    def sacar(self,valor):
        if valor < 0:
            return(False)
        if self.saldo < valor:
            return(False)

        self.saldo -= valor
        return(True)


    def depositar(self,valor):
        if valor < 0:
            return(False)
        self.saldo += valor
        return(True)


class ContaCorrente(Conta):
    def __init__(self, numero, agencia, cliente, limite, limite_saques):
        super().__init__(numero, cliente)
        self.limite = limite
        self.limite_saques = limite_saques

    def saque(self,valor):
        if valor > self.limite:
            return(False)
        saques = [transacao for transacoes in self.historico if transacao["transaction_type"] == Saque.__name__]
        num_saques = len(saques)
        if num_saques >= self.limite_saques:
            return(False)
        super().sacar(valor)
        return(True)
```
